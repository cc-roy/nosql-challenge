```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print("Number of establishments with a hygiene score equal to 20:", count)
# Display the first document in the results using pprint
first_document = establishments.find_one(query)
print("First document with a hygiene score equal to 20:")
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(establishments.find(query)))
# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:", len(df))
# Display the first 10 rows of the DataFrame
print("First 10 rows:")
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
pattern = "London"
query = {
    "LocalAuthorityName": {"$regex": pattern, "$options": "i"},  # Case-insensitive regex match for London
    "RatingValue": {"$gte": 4}  # RatingValue greater than or equal to 4
}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print("Number of establishments in London with RatingValue >= 4:", count)
# Display the first document in the results using pprint
first_document = establishments.find_one(query)
print("First document in London with RatingValue >= 4:")
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(establishments.find(query)))
# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:", len(df))
# Display the first 10 rows of the DataFrame
print("First 10 rows:")
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

penang_flavours = establishments.find_one({"BusinessName": "Penang Flavours"})
penang_geocode = penang_flavours["geocode"]
penang_latitude = float(penang_geocode.get("latitude", 0))
penang_longitude = float(penang_geocode.get("longitude", 0))

query = {
    "RatingValue": 5,
    "scores.Hygiene": {"$ne": ""}  # Exclude documents with empty hygiene scores
}

projection = {
    "_id": 0,
    "BusinessName": 1,
    "geocode.latitude": 1,
    "geocode.longitude": 1,
    "scores.Hygiene": 1
}

sort = [("scores.Hygiene", 1)]
limit = 5

cursor = establishments.find(query, projection).sort(sort).limit(limit)

df = pd.DataFrame(list(cursor))

df["Distance_to_Penang_Flavours"] = ((df["geocode"].apply(lambda x: x.get("latitude", 0)) - penang_latitude) ** 2 + 
                                              (df["geocode"].apply(lambda x: x.get("longitude", 0)) - penang_longitude) ** 2) ** 0.5

df.sort_values(by="Distance_to_Penang_Flavours", inplace=True)

print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score and nearest to 'Penang Flavours':")
print(df)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
query = {
    "scores.Hygiene": 0
}

# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": query},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},  # Sort from highest to lowest count
    {"$limit": 10}  # Limit to top 10 results
]

results = establishments.aggregate(pipeline)
# Print the number of documents in the result
total_documents = 0

for result in results:
    total_documents += result["count"]

print("Total number of documents in the results:", total_documents)
# Print the first 10 results
results = establishments.aggregate(pipeline)
print("Top 10 Local Authority areas with establishments having a hygiene score of 0:")
for result in results:
    print(result["_id"], ":", result["count"])
```

```python
results = establishments.aggregate(pipeline)
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(list(results))
# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:", len(df))
# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
df.head(10)
```
